### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset

We'll be using seaborn's iris dataset and a plot that we have already plotted in the theory in this example.

```python
iris = sns.load_dataset('iris')
iris['species'].replace({'setosa':'Iris Setosa', 'virginica':'Iris Virginica', 'versicolor':'Iris Versicolor'}, inplace = True)
iris.sample(5)
```

```python
iris.info()
```

### Plotting

```python
#Plotting a pairplot using seaborn
g = sns.pairplot(iris, hue = 'species')

#Setting the title
g.fig.suptitle("Pairplot on Iris dataset", fontsize = 'x-large', fontweight = 'bold', x = 0.5, y = 1.03)
g.fig.set_size_inches(10,7)

#Modifying the axes labels
labels = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]
for i in range(4):
    g.axes[i][0].set_ylabel(labels[i], fontsize = 'large', fontweight = 'bold')
    g.axes[3][i].set_xlabel(labels[i], fontsize = 'large', fontweight = 'bold')

#Modifying legend
legend = g.fig.get_children()[-1]
legend.set_title("Species", prop = {"size": "large", "weight" : 'bold'})
legend.set_bbox_to_anchor((1,0.97))
```

### Insights
These are just some insights that you can draw from this plot.
<ul>
    <li>The petal length and petal width features can be used to distinguish iris setosa from the other two species as they can be clearly seen as different clusters in the plots involving these features.</li>
    <li>Petal Length and Petal Width seem to have an almost linear correlation in all three flower species.</li>
    <li>Petal Width can act as a good feature to separate iris versicolor and iris virginica</li>
</ul>
